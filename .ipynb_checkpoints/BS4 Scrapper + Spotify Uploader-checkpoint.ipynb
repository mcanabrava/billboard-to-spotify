```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import configparser
import spotipy
from spotipy.oauth2 import SpotifyOAuth
```

```python
date = 1995 # use the year you want to get data from

response = requests.get("http://billboardtop100of.com/"+str(date)+"-2/")

print(response)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
# Create empty lists to store the data
ids = []
bands = []
songs = []

# Find all the rows (tr) in the table
rows = soup.find_all('tr')

# Loop through each row and extract the data from the columns (td)
for row in rows:
    cols = row.find_all('td')
    if len(cols) == 3:  # Ensure there are three columns (td) in the row
        ids.append(cols[0].text.strip())
        bands.append(cols[1].text.strip())
        songs.append(cols[2].text.strip())

# Create the DataFrame
data = {'id': ids, 'band': bands, 'song': songs}
song_df = pd.DataFrame(data)

song_df.head(10)
```

```python
# Read the Spotify API credentials and playlist name from the configuration file
config = configparser.ConfigParser()
config.read('spotify.cfg')

CLIENT_ID = config.get('SPOTIFY', 'CLIENT_ID')
CLIENT_SECRET = config.get('SPOTIFY', 'CLIENT_SECRET')
PLAYLIST_NAME = config.get('SPOTIFY', 'PLAYLIST_NAME')
USER_ID = config.get('SPOTIFY', 'USER_ID')
REDIRECT_URI='http://example.com'

client_id = CLIENT_ID
client_secret = CLIENT_SECRET
playlist_name = PLAYLIST_NAME
user_id = USER_ID

SCOPES = ['playlist-modify-public', 'playlist-modify-private', 'playlist-read-collaborative']
```

```python
# Step 1: Redirect the user for authorization
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=CLIENT_ID,
                                               client_secret=CLIENT_SECRET,
                                               redirect_uri=REDIRECT_URI,
                                               scope=SCOPES))
```

```python
# Step 1: Create a playlist
def create_playlist(user_id, playlist_name):
    playlist = sp.user_playlist_create(user=user_id, name=playlist_name, public=True)
    return playlist['id']

# Step 2: Add a track to the playlist
def add_track_to_playlist(playlist_id, track_uri):
    sp.user_playlist_add_tracks(user=sp.me()['id'], playlist_id=playlist_id, tracks=[track_uri])

# Step 3: Define search_track function using spotipy
def search_track(song_name, artist_name):
    query = f"{song_name} artist:{artist_name}"
    results = sp.search(q=query, type='track')
    track_uris = []
    for track in results['tracks']['items']:
        track_uris.append(track['uri'])
    return track_uris

# Step 4: Create custom playlist and add songs
def create_custom_playlist(song_df):

    # Step 5: Create the playlist
    playlist_id = create_playlist(user_id, playlist_name)

    # Step 6: Search for each song and add it to the playlist
    for index, row in song_df.iterrows():
        track_uris = search_track(row['song'], row['band'])
        for track_uri in track_uris:
            # Step 5: Add the track to the playlist
            add_track_to_playlist(playlist_id, track_uri)
            # Uncomment the following line to print the added song details
            print(f"Added: {row['song']} by {row['band']}")

# Assuming 'df' is the DataFrame containing the songs
create_custom_playlist(song_df)
```
